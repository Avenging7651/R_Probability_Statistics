```r
# データのロード
math <- read.csv("./csv/Math&Phys.csv", header = T)
phys <- math$phys
math <- math$math
```

```r
# 切断効果
#"データを切断して相関係数が下がり切断効果があることを確認する"
"全体の相関係数"
cor(math, phys)
"60点未満で切断"
cor(math[math<60], phys[math<60])
"60点以上で切断"
cor(math[math>=60], phys[math>=60])
"複雑なパターンでも見られる"
cor(math[math+phys>=120], phys[math+phys>=120])
```

```r
# 外れ値の影響
x <- rnorm(20,30,10)
y <- rnorm(20,30,10)
"相関係数"
cor(x,y)
"外れ地を追加する"
xa <- c(x,100)
ya <- c(y,100)
cor(xa,ya)
"大きく相関係数が上がっていることがわかる"
```
