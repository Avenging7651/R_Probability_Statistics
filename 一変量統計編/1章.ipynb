```r
# 1-3
# 34,56,32,15,49の 平均偏差と標準偏差
mean.deviation <- function(x){
    mean.x <- mean(x)
    1/length(x) * sum(abs(x - mean.x))
    }
standard.deviation <- function(x){
    sqrt(var(x)*(length(x)-1) / length(x))
}
data <-c(34,56,32,15,49)
mean.deviation(data)
standard.deviation(data)
```

```r
# 1-4
# 平均点65, 59, 62, 人数 500, 750, 690 人 全体の平均点を求めよ
a <- c (65,59,62)
b <- c(500,750,690)
sum(a * b)/sum(b)
```

```r
# 1-5
# 171.8cm 74.4kg, 167,2cm 56.3kg, 180.9cm 93.2kg のBMIを求めよ
# ただしheight, weighといういうベクトルとして計算せよ
BMI <- function(h_cm,w_kg){
    w_kg / (0.01 * h_cm)^2
}
height <- c(171.8,167.2,180.9)
weight <- c(74.4,56.3,93.2)
BMI(height,weight)
```

```r
# 1-6
# 3,4,8,11, 7の不偏分散の平方根と標準偏差を求めよ
data <- c(3,4,8,11, 7)
sd(data)
standard.deviation(data)
```

```r
# 1-7
# z-値を,偏差値を50 + 10Z とし, 40点,点の場合の偏差値をそれぞれ求めよ
Z.score <- function(x, samples){
    standard.deviation <- sqrt(var(samples)*(length(samples)-1) / length(samples))
    (x - mean(samples)) / standard.deviation
}
50 + 10 * Z.score(c(40,85), math$math)
```

```r
# 1-12
skewness <- function(x,k){
    standard.deviation <- sqrt(var(x)*(length(x)-1) / length(x))
    z.score <- (x - mean(x)) / standard.deviation
    mean(z.score ^ k)
}
kurtosis <- function(x){
    skewness(x,4) -3
}
skewness(math$math,3)
kurtosis(math$math)
```

xの分散σ^2がx, x^2のそれぞれの平均を用いてσ^2 = x^2の平均　- (xの平均)^2であることを示せ
$$
\begin{align}
\sigma ^2 &=\frac{\sum{(x_i - \bar{x})}}{n} \\
&= \frac{1}{n}\sum{x_j^2}-\frac{2}{n}\bar{x}\sum{x_j} + \frac{n}{n}\bar{x}^2 \\
&=\frac{1}{n}\sum{x_j^2}- 2\bar{x}^2 + \bar{x}^2\\
&=\bar{x^2} - \bar{x}^2
\end{align}
$$
